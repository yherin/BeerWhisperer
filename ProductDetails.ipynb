```python
import pandas as pd
from bs4 import BeautifulSoup

products = pd.read_csv('./beers.csv', dtype={'Numero': object} )
for p in products['Numero']:
    print(p)
    with open(f'./html/{p}.html', encoding="utf-8") as fp:
        try:
            soup = BeautifulSoup(fp, 'html.parser')
            print(soup.title.string)
            print(soup.find_all("li", attrs={"class": "column"}))
            print(soup.find_all("div", attrs={"class": "html-print"}))
            #print(soup.find_all("div", attrs={"class":"pdp main clearfix black js-accessibility-main-content"}))
        except UnicodeDecodeError:
            print("error")
            pass
```
